```python
#[99,86,87,88,111,94,78,77,85,86,86]
#Numerical 1,2,3,0
#Categorical ex :colors
#Ordinal ex: A,B ,C grades
import numpy
x= [99,86,87,88,111,94,78,77,85,86,86]
mean= numpy.mean(x)
print('MEAN=' ,mean)

median =numpy.median(x)
print('Median=',median)


from scipy import stats
mode = stats.mode(x)
print('Mode=',mode)


standard_deviation= numpy.std(x)
print('STD=',standard_deviation)

var=standard_deviation**2
variance= numpy.var(x)
print('Variance=',variance)
print('Variance=',var)

    


```

```python
percentile=numpy.percentile(x,0)
```
